```python
import pywt

def denoise_ecg(signal, wavelet='sym5', level=5):
    """Denoise ECG signal using Discrete Wavelet Transform (DWT)."""
    coeffs = pywt.wavedec(signal, wavelet, level=level)  # Wavelet decomposition
    coeffs[1:] = [pywt.threshold(c, np.std(c), mode='soft') for c in coeffs[1:]]  # Soft thresholding
    denoised_signal = pywt.waverec(coeffs, wavelet)  # Reconstruct signal
    return denoised_signal[:len(signal)]  # Ensure length consistency
```

```python
# Apply DWT denoising
denoised_ecg = denoise_ecg(ecg_signal)
```

```python
# Plot the original and denoised signals
plt.figure(figsize=(12, 5))
plt.plot(ecg_signal, color='blue')
plt.title("Original ECG Signal")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(denoised_ecg, color='red')
plt.title("Denoised ECG Signal using DWT")
plt.xlabel("Time (samples)")
plt.ylabel("Amplitude")
plt.show()
```
