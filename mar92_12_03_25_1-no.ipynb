```python
ann_symbols = ann.symbol
ann_symbols
```

```python
ann_samples = ann.sample
ann_samples
```

```python
ann_aux = ann.aux_note
ann_aux
```

```python
ann.__dict__
```

```python
vf_index = None
for i, aux in enumerate(ann_aux):
    if aux and "VF" in aux:
        vf_index = ann_samples[i]
        break
```

```python
vf_index
```

```python
if vf_index is None:
    print("VF onset not found in aux notes, checking symbol codes...")
    # (In some databases, VF might be encoded as a symbol, e.g., "[" for VF start)
    for i, sym in enumerate(ann_symbols):
        if sym == "[":  # assuming "[" might mark VF start in this annotation set
            vf_index = ann_samples[i]
            break
```
